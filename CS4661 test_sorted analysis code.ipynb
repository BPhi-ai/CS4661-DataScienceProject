```python
#0 = no value, #1 = correct, #2 = incorrect #3 = timed-out #4 = skipped.
df['correct?'] = outcome['correct']
df['status of question'] = outcome['outcome']
df
```

```python
#for every question, how many users have encountered that question.
usersPerQuestion = []
usersPerQuestionWithoutZero = []
questionIDWithoutZeroUsers = []

for n in range(5477):
    x = df.loc[df['question_id'] == n]
    if(len(x) != 0):
        usersPerQuestion.append(len(x))
        questionIDWithoutZeroUsers.append(df['question_id'].loc[df.index[n]])
        print('question id: ' + str(df['question_id'].loc[df.index[n]]) + " users: " + str(len(x)))

for x in usersPerQuestion:
    if (x != 0):
        usersPerQuestionWithoutZero.append(x)

for y in usersPerQuestionWithoutZero:
    print("users: " + str(y))
    
#list of question_id that has at least 1 user.
noduplicate = list(set(questionIDWithoutZeroUsers))
```

```python
for a in noduplicate:
    print("question_id: " + str(a))
```
