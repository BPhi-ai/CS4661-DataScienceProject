### Getting a list of tag strings for all the incorrect answers from user 1
Each question has a tag string that corresponds to what category that question falls into. This will be useful later

```python
testList = []
for tag in df_user_1_incorrect['tag_string']:
    testList.append(tag)
testList2 = []

for i in testList:
    for j in i.split():
        testList2.append(j)
```

```python
testList2
```

```python
user1_dict = {}
```

```python
for i in testList2:
    if i in user1_dict:
        val = user1_dict.get(i)
        user1_dict.update({i:val+1})
    else:
        user1_dict.update({i:1})
```
